**Basic Concept**
An eigenvector is a nonzero vector that, when transformed by a matrix, only changes in scale (magnitude) but not in direction. If vector v is an eigenvector of matrix A, then:

$$ Av = \lambda v $$

where:
- A is a square matrix.
- v is the eigenvector.
- λ is the eigenvalue.

**Geometric Interpretation**
- When a linear transformation is applied, eigenvectors maintain their direction
- The eigenvalue indicates how much the vector is scaled
- Negative eigenvalues reverse the direction
- Zero eigenvalues collapse the vector to zero


### Step-by-Step for Eigenvalues and Eigenvectors Calculation

We are working with the matrix:

$$
A = \begin{bmatrix} 2 & 0 \\ 0 & 3 \end{bmatrix}
$$


### **Step 1: Find the Eigenvalues**

The eigenvalues are determined by solving the **characteristic equation**:

$$
\det(A - \lambda I) = 0
$$

1. Subtract λI (where I is the identity matrix) from A:
   $$
   A - \lambda I = 
   \begin{bmatrix} 
   2 - \lambda & 0 \\ 
   0 & 3 - \lambda 
   \end{bmatrix}
   $$

2. Compute the determinant:
   $$
   \det(A - \lambda I) = (2 - \lambda)(3 - \lambda) - (0)(0)
   $$
   Simplify:
   $$
   (2 - \lambda)(3 - \lambda) = 0
   $$

3. Solve for λ:
   $$
   (2 - \lambda) = 0 \quad \text{or} \quad (3 - \lambda) = 0
   $$
   This gives:
   $$
   \lambda_1 = 2, \quad \lambda_2 = 3
   $$

## Singular Value Decomposition (SVD)

**Definition**
SVD is a matrix factorisation technique that decomposes any matrix A into three components:

$$ A = U\Sigma V^T $$

where:
- U is an orthonormal matrix containing left singular vectors
- Σ (Sigma) is a diagonal matrix containing singular values
- V^T is the transpose of an orthonormal matrix containing right singular vectors

**Key Properties**
- Singular values are always non-negative real numbers
- They are arranged in descending order
- The number of non-zero singular values equals the rank of the matrix

## Mathematical Advantages

**1. Stability**
- SVD is numerically stable
- Less sensitive to rounding errors
- Provides optimal low-rank approximations

**2. Uniqueness**
- Singular values are unique
- Provides consistent decomposition
- Independent of matrix orientation

This decomposition serves as a fundamental tool in various fields, from quantum mechanics to facial recognition, making it one of the most important matrix factorization techniques in linear algebra and its applications.

## SVD Matrix Visualization Example

**Original Matrix A (3×3)**
$$
A = \begin{bmatrix} 
4 & 0 & 3 \\
0 & -2 & 0 \\
3 & 0 & 4
\end{bmatrix}
$$

## Step-by-Step Decomposition

**1. Finding Σ (Singular Values)**
- First calculate
$$
A^TA = \begin{bmatrix}
25 & 0 & 24 \\
0 & 4 & 0 \\
24 & 0 & 25
\end{bmatrix}
$$

- Find eigenvalues (λ) by solving |A^TA - λI| = 0
- Singular values (σ) are square roots of these eigenvalues
$$
\Sigma = \begin{bmatrix}
7 & 0 & 0 \\
0 & 2 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

**2. Finding V (Right Singular Vectors)**
- Eigenvectors of A^TA form columns of V
$$
V = \begin{bmatrix}
0.707 & 0 & 0.707 \\
0 & 1 & 0 \\
0.707 & 0 & -0.707
\end{bmatrix}
$$

**3. Finding U (Left Singular Vectors)**
- U = AV/σ for each singular value
$$
U = \begin{bmatrix}
0.707 & 0 & 0.707 \\
0 & -1 & 0 \\
0.707 & 0 & -0.707
\end{bmatrix}
$$

## Visual Representation of SVD

```

(3×3)         (3×3)              (3×3)              (3×3)

[4 0 3]   [0.707  0  0.707]   [7 0 0]   [0.707  0   0.707]
[0 -2 0] = [ 0   -1   0   ] × [0 2 0] × [ 0    1    0   ]
[3 0 4]   [0.707  0 -0.707]   [0 0 1]   [0.707  0  -0.707]
```

## Properties Visualization

**1. Orthogonality Check**
- For U: U^T × U = I
- For V: V^T × V = I

**2. Rank Visualization**
- Number of non-zero singular values = rank of matrix
- In this case, rank = 3 (full rank)

**3. Matrix Reconstruction**
```python
# Reconstruct using k singular values
k = 2  # Using top 2 singular values

A_k = U[:,:k] × Σ[0:k,0:k] × V^T[0:k,:]
        3x2        2x2           2x3  
```

**4. Energy Conservation**
- Total energy = sum of squares of singular values
$$
E_{total} = 7^2 + 2^2 + 1^2 = 54
$$

````python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load image and convert to grayscale
img = Image.open('media/9420.JPG').convert('L')
img_array = np.array(img)

# Display original image
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img_array, cmap='gray')
plt.axis('off')
````

````python
img_small = img.resize((10, 10))  # Small 10x10 image for matrix visualization
img_array_small = np.array(img_small)

# # Display original small image
# plt.figure(figsize=(12, 12))
# plt.subplot(2, 1, 1)
# plt.title('10x10 Grayscale Image')
# plt.imshow(img_array, cmap='gray')
# plt.axis('off')

# Display image as matrix with values
plt.figure(figsize=(12, 12))
plt.subplot(2, 1, 2)
plt.title('Image Matrix (Pixel Values)')
plt.imshow(img_array_small, cmap='gray')
for i in range(img_array_small.shape[0]):
    for j in range(img_array_small.shape[1]):
        plt.text(j, i, f'{img_array_small[i, j]}', 
                ha='center', va='center', color='red')
plt.colorbar()
plt.show()
````

````python
# Decompose image matrix into U, S, and Vt
U, S, Vt = np.linalg.svd(img_array, full_matrices=False)

print(f"Shape of U: {U.shape}")
print(f"Shape of S: {S.shape}")
print(f"Shape of Vt: {Vt.shape}")
print(f"\nFirst few singular values: {S[:100]}")
````

````python
# Keep only top 100 singular values
k =100
S_compressed = np.zeros((k))
S_compressed[:k] = S[:k]

# Reconstruct image using compressed components
compressed_img = np.dot(U[:, :k], np.dot(np.diag(S_compressed), Vt[:k, :]))

# Display compressed image
plt.subplot(1, 2, 2)
plt.title(f'Compressed Image (k={k})')
plt.imshow(compressed_img, cmap='gray')
plt.axis('off')
plt.show()
````

````python
# Perform SVD
U, S, Vt = np.linalg.svd(img_array, full_matrices=False)

# Calculate cumulative energy (variance) percentage
total_energy = np.sum(S**2)
cumulative_energy = np.cumsum(S**2) / total_energy * 100

# Find number of singular values needed for 99% energy
k_99 = np.where(cumulative_energy >= 99)[0][0] + 1

print(f"Total number of singular values: {len(S)}")
print(f"Number of singular values for 99% energy: {k_99}")
````

````python
# Plot singular values
plt.figure(figsize=(12, 4))

# Singular values plot
plt.subplot(1, 2, 1)
plt.plot(S, 'bo-')
plt.title('Singular Values')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.grid(True)

# Cumulative energy plot
plt.subplot(1, 2, 2)
plt.plot(cumulative_energy, 'ro-')
plt.axhline(y=99, color='g', linestyle='--', label='99% threshold')
plt.axvline(x=k_99, color='r', linestyle='--', label=f'k={k_99}')
plt.title('Cumulative Energy %')
plt.xlabel('Number of Singular Values')
plt.ylabel('Cumulative Energy %')
plt.grid(True)
plt.legend()
plt.show()
````

### Compression Ratio and comparison

````python
original_size = img_array.size
compressed_size = k * (U.shape[0] + Vt.shape[1] + 1)
compression_ratio = original_size / compressed_size

print(f"Original size: {original_size} elements")
print(f"Compressed size: {compressed_size} elements")
print(f"Compression ratio: {compression_ratio:.2f}:1")
````

````python
# Calculate and display reconstruction error
error = np.abs(img_array - compressed_img)
plt.figure(figsize=(8, 4))
plt.title('Reconstruction Error')
plt.imshow(error, cmap='hot')
plt.colorbar()
plt.axis('off')
plt.show()

# Print mean squared error
mse = np.mean((img_array - compressed_img) ** 2)
print(f"Mean Squared Error: {mse:.2f}")
````

## Understanding Mean Squared Error (MSE) in Image Compression

**Basic Formula**
$$ MSE = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2 $$

where:

- Y_i = Original pixel value
- \hat{Y}_i = Compressed pixel value
- n = Total number of pixels


Let's analyze a 3×3 image patch:

**Original Image Patch**
```
Original (Y)
[120  125  128]
[115  118  122]
[110  112  115]
```

**Compressed Image Patch**
```
Compressed (Ŷ)
[118  123  125]
[114  116  120]
[108  110  113]
```

**Difference Matrix (Y - Ŷ)**
```
[ 2   2   3]
[ 1   2   2]
[ 2   2   2]
```

## MSE Calculation Steps

**1. Square the Differences**
```
[4  4   9]
[1  4   4]
[4  4   4]
```

**2. Sum All Values**
```
4 + 4 + 9 + 1 + 4 + 4 + 4 + 4 + 4 = 38
```

**3. Divide by Number of Pixels**
```
MSE = 38/9 ≈ 8.43
```
